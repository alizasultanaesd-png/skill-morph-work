```python
# Converting 'yes' to 1 and 'no' to 0
df['mainroad'] = df['mainroad'].map({'yes': 1, 'no': 0})
print("Converted mainroad values:", df['mainroad'].head())
```

```python
# Split into Train and Test
# TODO: Split - 80% train, 20% test

# Assuming 'df' is full dataset and predicting 'price'
X = df.drop('price', axis=1)
y = df['price']

# Spliting into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

from sklearn.linear_model import LogisticRegression

# Create model
model = LogisticRegression(max_iter=2000, solver='lbfgs')

# Train the model
model.fit(X_train, y_train)
print("Model trained!")

# Check accuracy on training data
train_score = model.score(X_train, y_train)
print(f"Training accuracy: {train_score:.1%}")

# Check accuracy on test data
test_score = model.score(X_test, y_test)
print(f"Test accuracy: {test_score:.1%}")
```

```python
from sklearn.model_selection import cross_val_score

# Do 5-fold cross-validation
scores = cross_val_score(
    model,           # Our model
    X_train,  # Training data
    y_train,         # Training labels
    cv=5            # 5 mini-tests
)

print("5 mini-test scores:")
for i in range(5):
    print(f"  Test {i+1}: {scores[i]:.1%}")

print(f"\nAverage: {scores.mean():.1%}")
print(f"This means our model is {scores.mean():.1%} accurate!")
```
