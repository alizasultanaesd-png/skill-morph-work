# **Bar Chart**

```python
# 1. Bar chart of gym membership types
import pandas as pd
member = pd.read_csv('gym_membership.csv')
member_count = member['abonoment_type'].value_counts()
plt.bar(member_count.index, member_count.values, color=['pink', 'blue'])
plt.title('Gym Membership Types')
plt.xlabel('Membership Type')
plt.ylabel('Number of Members')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
#plt.savefig("/content/drive/MyDrive/datasets/colorbarchartgendervalue.png",dpi=300, bbox_inches='tight')
plt.show()

```

# **Scatter Plot**

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

plt.figure(figsize=(8, 5))
plt.scatter(gym['Age'], gym['visit_per_week'], color='teal', alpha=0.6, edgecolors='w')
plt.title("Age vs Weekly Gym Visits")
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# **Histogram**

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])

# Math score distribution
plt.hist(gym['avg_time_in_gym'], bins=15, color='mediumseagreen', edgecolor='black')
plt.title("Distribution of Average Time Spent in Gym")
plt.xlabel("Time in Gym (minutes)")
plt.ylabel("Number of Members")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# **Multiple Plots**

```python
# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
sns.set_style("whitegrid")

# Plot 1: Age distribution (Histogram)
sns.histplot(gym['Age'], bins=15, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Age Distribution")
axes[0, 0].set_xlabel("Age")
axes[0, 0].set_ylabel("Count")

# Plot 2: Gender count (Countplot)
sns.countplot(data=gym, x='gender', ax=axes[0, 1], palette='pastel')
axes[0, 1].set_title("Gender Count")
axes[0, 1].set_xlabel("Gender")
axes[0, 1].set_ylabel("Number of Members")

# Plot 3: Visits per week (Histogram)
sns.histplot(gym['visit_per_week'], bins=10, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title("Visits per Week")
axes[1, 0].set_xlabel("Visits")
axes[1, 0].set_ylabel("Count")

# Plot 4: Personal training (Countplot)
sns.countplot(data=gym, x='personal_training', ax=axes[1, 1], palette='coolwarm')
axes[1, 1].set_title("Personal Training Participation")
axes[1, 1].set_xlabel("Personal Training")
axes[1, 1].set_ylabel("Number of Members")

plt.tight_layout()
#plt.savefig("/content/drive/MyDrive/datasets/multipleplot.png",dpi=300)
plt.show()
```

# **Seaborn**

# **Box Plot**

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

plt.figure(figsize=(7, 5))
sns.boxplot(data=gym, x='gender', y='Age')
plt.title("Age Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# **Count Plot**

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')


sns.countplot(data=gym, x='days_per_week')
sns.countplot(data=gym, x='days_per_week', palette='coolwarm')
plt.title("Frequency of Gym Visits per Week")
plt.xlabel("Days per Week")
plt.ylabel("Number of Members")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# **Bar Plot with Average**

```python
# 3. Bar plot: Average visits by membership type # Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')

sns.barplot(data=gym, x='abonoment_type', y='visit_per_week', palette='viridis')
plt.title("Average Weekly Visits by Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Average Visits per Week")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
